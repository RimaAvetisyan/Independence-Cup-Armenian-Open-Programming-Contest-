```python
import cv2
import os

def add_padding(image, padding_size=10):
    """
    Adds white padding around the image.
    
    Parameters:
    - image: The input image.
    - padding_size: The width of the padding to add (default is 20 pixels).
    
    Returns:
    - Padded image.
    """
    return cv2.copyMakeBorder(image, padding_size, 0, padding_size, 0,
                              cv2.BORDER_CONSTANT, value=[255, 255, 255])  # White padding

                              
def split_and_save_images(image, rows, cols, output_folder):
    """
    Splits a large image into rows x cols smaller images (cards) and saves them as individual files.
    
    Parameters:
    - image: The large input image containing multiple cards.
    - rows: Number of rows of cards.
    - cols: Number of columns of cards.
    - output_folder: Directory to save the card images.
    
    The files will be saved in the format 'card_row_col.jpg'.
    """
    # Get image dimensions
    image = add_padding(image)
    img_height, img_width = image.shape[:2]
    
    # Calculate height and width of each card
    card_height = img_height // rows
    card_width = img_width // cols
    
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    card_count = 1
    
    # Loop over rows and columns to crop and save each card
    for row in range(rows):
        for col in range(cols):
            # Crop each card
            card = image[row * card_height : (row + 1) * card_height, 
                         col * card_width : (col + 1) * card_width]
            
            # Generate file name
            card_filename = f'card_{row+1}_{col+1}.png'
            card_path = os.path.join(output_folder, card_filename)
            
            # Save the card image
            cv2.imwrite(card_path, card)
            
            print(f"Saved {card_path}")
            card_count += 1

# Load the large image containing 35 cards
image = cv2.imread('images2/0001.png')
image2 = cv2.imread('images2/0002.png')

# Output folder where you want to save the card images
output_folder = 'card_templates/'
output_folder2 = 'card_templates2/'

# Split and save the large image (5 rows and 7 columns of cards)
split_and_save_images(image, 5, 7, output_folder)
split_and_save_images(image, 5, 7, output_folder2)
```

```python
import cv2
import os

def split_image(image, rows, cols):
    """
    Splits a large image into rows x cols smaller images (cards).
    
    Returns a list of individual card images.
    """
    # Get image dimensions
    img_height, img_width = image.shape[:2]
    
    # Calculate height and width of each card
    card_height = img_height // rows
    card_width = img_width // cols
    
    cards = []
    
    # Loop over the rows and columns and crop each card
    for row in range(rows):
        for col in range(cols):
            card = image[row * card_height : (row + 1) * card_height, 
                         col * card_width : (col + 1) * card_width]
            cards.append(card)
    
    return cards

def load_card_templates(template_folder):
    """
    Loads all card templates from a given folder and stores them in a dictionary.
    """
    card_templates = {}
    for filename in os.listdir(template_folder):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            template_path = os.path.join(template_folder, filename)
            card_image = cv2.imread(template_path)
            card_name = os.path.splitext(filename)[0]  # Use the filename without extension as key
            card_templates[card_name] = card_image
    return card_templates

def recognize_card(card_image, templates):
    """
    Recognizes a card using template matching. Returns the name of the recognized card.
    
    Parameters:
    - card_image: The image of the card to be recognized.
    - templates: A dictionary of template images for all cards.
    
    Returns:
    - The recognized card name.
    """
    best_match = None
    highest_score = 0
    
    for card_name, template in templates.items():
        # Resize the template to match the card size if needed (depending on your image size)
        template_resized = cv2.resize(template, (card_image.shape[1], card_image.shape[0]))
        
        # Template matching
        res = cv2.matchTemplate(card_image, template_resized, cv2.TM_CCOEFF_NORMED)
        _, score, _, _ = cv2.minMaxLoc(res)
        
        if score > highest_score:
            highest_score = score
            best_match = card_name
    
    return best_match

def find_missing_card(large_image, templates, rows, cols):
    """
    Splits the large image into individual cards, recognizes each card using templates,
    and identifies the missing card from the deck.
    
    Parameters:
    - large_image: The large image containing 35 cards.
    - templates: A dictionary of template images for all 36 cards.
    - rows: Number of rows of cards in the large image.
    - cols: Number of columns of cards in the large image.
    
    Returns:
    - The name of the missing card.
    """
    # Full deck of cards (36 cards)
    full_deck = set(templates.keys())
    
    # Step 1: Split the large image into individual cards
    cards_in_image = split_image(large_image, rows, cols)
    
    # Step 2: Recognize each card
    recognized_cards = []
    for card_image in cards_in_image:
        recognized_card = recognize_card(card_image, templates)
        recognized_cards.append(recognized_card)
    
    # Step 3: Find the missing card
    recognized_set = set(recognized_cards)
    missing_card = full_deck - recognized_set
    
    # Return the missing card (there should only be one missing)
    return missing_card.pop() if missing_card else None

def process_images_in_folder(image_folder, templates, rows, cols):
    """
    Reads all images in a folder, finds the missing card in each, and prints the results.
    
    Parameters:
    - image_folder: The folder containing images with 35 cards each.
    - templates: A dictionary of template images for all 36 cards.
    - rows: Number of rows of cards in each large image.
    - cols: Number of columns of cards in each large image.
    """
    for filename in os.listdir(image_folder):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(image_folder, filename)
            large_image = cv2.imread(image_path)
            
            if large_image is None:
                print(f"Could not read image {filename}")
                continue
            
            # Find the missing card in the current image
            large_image_padded = add_padding(large_image)
            missing_card = find_missing_card(large_image_padded, templates, rows, cols)
            print(missing_card)

# Folder containing images to process
image_folder = 'images2/'

# Folder containing card templates
template_folder = 'card_templates/'

# Load card templates
templates = load_card_templates(template_folder)

# Process all images in the folder and find the missing card in each
process_images_in_folder(image_folder, templates, 5, 7)

print("End")
```
